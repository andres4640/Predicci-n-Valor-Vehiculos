Creamos un nuevo dataset donde se realizara la limpieza de la columna

```python
dt_torque = dt.copy()
```

Creamos una función para saber en que forma se esta midiendo el torque. Esta columna se esta midiendo segun Newton-metro (Nw) o kilogramo-metro (kgm).

```python
def medida_valor(row):
    if "nm" in row or "Nm" in row or "NM" in row:
        return "nm"
    elif "KGM" in row or "kgm" in row:
        return "kgm"
```

```python
dt_torque["medida"] = dt_torque["torque"].apply(medida_valor)
```

Vemos que valores de la columna torque no estan medidos por newton-metro y kilogramo-metro. Estas filas pueden ser eliminadas ya que solo son 12.

```python
print(dt_torque[dt_torque["medida"].isna()]["torque"])
print("")
print("Numero de filas:", dt_torque["medida"].isna().sum())
```

```python
dt_torque = dt_torque.dropna().reset_index(drop=True)
```

Creamos una función para limpiar la columna "torque" y solo extraer el numero que lo representa. Este numero es el primero que aparece en la fila. 

```python
import re
def torque_numerico(row):
    numeros = re.findall("\d+\.\d+", row)
    if len(numeros) == 0:
        numeros = re.findall("\d+", row)
    return numeros[0]
```

```python
dt_torque["clean_torque"] = dt_torque["torque"].apply(torque_numerico).astype(float)
```

```python
dt_torque["clean_torque"].head(10)
```

Ahora que tenemops la columna "clean_torque" con los valores numericos, debemos uniformizar la medición de esta, ya que los valore se encuentran medidos en newton-metro y kilogramo-metro. Convertiremos los que se encuentran en kilogramo-metro a newton-metro.

Investigamos en google sobre la conversión de medidas: 1 kg-m es igual a 9.80665 newton-metro.

```python
def torque_uniforme(row):
    if row.medida == "kgm":
        return row.clean_torque * 9.80665
    else:
        return row.clean_torque
```

```python
dt_torque["clean_torque"] = dt_torque.apply(torque_uniforme, axis=1)
```

Calcularemos el rango intercuartil para eliminar los outliers. Estos valores pueden sesgar nuestro modelo de predicción y hacer que produzca resultados erroneos. 

Para calcular este rango restaremos el percentil 75 menos el percentil 25, luego calcularemos los limites superior e inferior y eliminaremos los valores que se encuentran afuera de este rango.

```python
def rango_intercuartil(ds, col):
    rango_intercuartil = ds[col].quantile(0.75) - ds[col].quantile(0.25)
    limite_superior = ds[col].quantile(0.75)  + rango_intercuartil*1.5
    limite_inferior = ds[col].quantile(0.25) - rango_intercuartil*1.5
    
    return (limite_inferior, limite_superior)
```

```python
lim = rango_intercuartil(dt, "selling_price")
print("Valor del Limite inferior ", lim[0])
print("Valor del Limite superior ", lim[1])
```

Hay 598 datos que son outliers 

```python
num_filas_outliers = len(dt[dt["selling_price"] >= lim[1]])
print("Numero de datos outliers:", num_filas_outliers)

porcentaje_filas_outliers = num_filas_outliers / dt.shape[0] *100
print("Porcentaje datos outliers:", porcentaje_filas_outliers)
```

## ANOVA

Debemos validar los supuestos del analisis ANOVA:

1. Datos estan normalmente Distribuidos
2. Homogendidad entre las varianzas de los grupos.

### Supuestos

Primero veremos si los datos estan distribuidos normalmente.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
from statsmodels.graphics.gofplots import qqplot
import pylab
```

Creamos un modelo OLS por cada variable categorica.

```python
model = ols('selling_price ~ C(fuel)', dt_no_out).fit()
print(model.summary())

model2 = ols('selling_price ~ C(seller_type)', dt_no_out).fit()
print(model2.summary())

model3 = ols('selling_price ~ C(transmission)', dt_no_out).fit()
print(model3.summary())

model4 = ols('selling_price ~ C(owner)', dt_no_out).fit()
print(model4.summary())
```

#### Shapiro Test

Aplicamos el Shapiro Test a cada variable categorica para saber si tienen una distribución normal. Si el P-Value es mayor a 0.05, entonces los grupos de la variable categorica tienen una distribución normal.

```python
w, pvalue = stats.shapiro(model.resid)
print('Fuel: Statistics=%.3f, p=%.3f' % (w, pvalue))

w, pvalue = stats.shapiro(model2.resid)
print('seller_type: Statistics=%.3f, p=%.3f' % (w, pvalue))
      
w, pvalue = stats.shapiro(model3.resid)
print('transmission: Statistics=%.3f, p=%.3f' % (w, pvalue))
      
w, pvalue = stats.shapiro(model4.resid)
print('owner: Statistics=%.3f, p=%.3f' % (w, pvalue))
```

#### QQ-Plot

```python
sm.qqplot(model.resid, dist="norm")
plt.show()

sm.qqplot(model2.resid, dist="norm")
plt.show()

sm.qqplot(model3.resid, dist="norm")
plt.show()

sm.qqplot(model4.resid, dist="norm")
plt.show()
```

A los QQ-Plot y la prueba Shapiro, podemos determinar que las distribuciónes de los diferentes grupos en las variables categoricas no son normales.

Aplicaremos la prueba Kruskal-Wallis para determinar si hay una diferencia estadisticamente significativa entre los grupos de las variables categoricas en la variable dependiente.

##  Kruskal-Wallis Test

```python
from pingouin import kruskal
```

```python
kruskal(data=dt_no_out, dv='selling_price', between="fuel")
```

```python
kruskal(data=dt_no_out, dv='selling_price', between="seller_type")
```

```python
kruskal(data=dt_no_out, dv='selling_price', between="transmission")
```

```python
kruskal(data=dt_no_out, dv='selling_price', between="owner")
```

En base a la prueba Kruskal, podemos determinar que por lo menos dos grupos de cada variable categorica son diferentes de forma estadisticamente significativo respecto a la variable dependiente, precio del vehiculo.
